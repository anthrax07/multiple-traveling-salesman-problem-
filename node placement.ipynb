```python
import random
import random
import networkx as nx
import math
import itertools    
import matplotlib.pyplot as plt

```

```python
#this class is created to define the position of the sensor and possible drone points

class Node:
    posX=0  
    posY=0 
    data_rate=50 #data transmission rate for both drone and sensor
    transmission_range=200
    def __init__(self):
        self.posX
        self.posY
        self.data_rate
        self.transmission_range
   
    def get_pose_node(self):
        self.posX=random.randrange(500)
        self.posY=random.randrange(500)
        self.data_rate
        self.transmission_range
        return self.posX,self.posY
    
```

```python
sensor=[]
```

```python
#here to get the sensor node points I have made a sensor list where sensor objects are stored

number_of_sensor=20
for i in range(number_of_sensor):
    
    sensor.append(Node().get_pose_node())
    
```

```python
#function to calculate distance between two drone positions
def distance(points, i, j):
    x=points[i][0]-points[j][0]
    y=points[i][1]-points[j][1]
    return math.sqrt(x*x + y*y)
```

```python
G=nx.empty_graph()
```

```python
for u in range(len(drone)):
    if u<len(drone)-1:
        v=u+1
        wt=distance(drone,u,v)
        print(u,v)
        G.add_edge(u,v,weight=wt)
```

```python
nx.draw_networkx(G,drone)
```

```python
edges,weights=zip(*nx.get_edge_attributes(G,'weight').items())
```
